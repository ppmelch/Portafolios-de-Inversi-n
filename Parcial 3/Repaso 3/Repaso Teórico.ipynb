## **Rebalanceos**

### Rebalanceo de Portafolios

- Proceso en el que se ajustan las ponderaciones de los activos que conforman una cartera de inversión para mantenerla en linea con los objetivos y la estrategia de inversión establecidos.

- Consiste en vender o comprar activos en la cartera para establecer las ponderaciones objetivo , que se desvían debido a las fluctuaciones del mercado.



- Un rebalanceo se realiza periódicamente de acuerdo a intervalos predefinidos , puede ser trimestral , semestral o anual.

- Cuando se realiza el rebalanceo , se venden los activos que han aumentado en valor y se compran los activos que han disminuido en valor, de manera que se ajusta la distribución de activos de la cartera a los porcentajes objetivo.

- Esto implica vender los activos que han tenido un mejor desempeño y comprar los activos que han tenido un rendimiento inferior.

- Un rebalanceo también puede implicar cambiar las ponderaciones , re-optimizando el portafolio.

### Overweight 

- Cuando un activo "pesa" por arriba de su ponderación objetivo debido a su fluctuación positiva del mismo , se dice que el portafolio esta overweight en ese activo. Ya que su ponderación real en el portafolio es mayor que la objetivo. 

### Underweight 

- Cuando un activo "pesa" por debajo de su ponderación objetivo debido a la fluctuación negativa del mismo , se dice que el portafolio esta underweight en ese activo. Ya que su ponderación real en el portafolio es menor a la objetivo.

### Beneficios 
#### Control del Riesgo

- El rebalanceo ayuda a controlar el riesgo al garantizar que la cartera no se concentre en activos que han tenido un buen desempeño reciente.

#### Oportunidad

- El rebalanceo permite aprovechar oportunidades de compra cuando los activos han disminuido en valor (infravalorados).

#### Estrategia 
- El rebalanceo ayuda a mantener la estrategia de inversión a largo plazo del inversor. Si se permitiera que la cartera se desviara , se podróa alterar la exposición al riesgo y los objetivos de inversión establecidos.

### Tipos de Fluctuaciones que afectan los pesos de un portafolio
#### Fluctuaciones de Mercado
- Fluctuaciones generales del mercado de valores donde existen movimientos de los precios de las acciones y otros activos financieros.

- Estas fluctuaciones pueden estar influenciadas por factores macroeconómicos , eventos geopolíticos , noticias económicas y financieras, cambios en las tasas de interés , condiciones económicas generales, entre otros.

#### Fluctuaciones de Tipo de Cambio
- Las fluctuacioens de tipo de cambio se refieren a los cambios en el valor de una monerda en relación con otra. Estas fluctuaciones tienen un impacto alto en las acciones que cotizan en el SIC.

    - SIC: Sistema Internacional de Cotizaciones de Bolsa de Mexicana de Valores (BMV) que permite a los inversionistas mexicanos acceder a activos listados en bolsas de valores extranjeras.

- Las bajas de las monedas de referencias contra el peso (USD/MXN , EUR/MXN) afectan en gran medida a los administradores de portafolios que compran en el SIC, ya que , la fluctuación del tipo de cambio se puede "comer" la ganancia de un activo.
#### Fluctuaciones Sectoriales

- Los diferentes sectores de la economía pueden experimentar fluctuaciones de precios debido a factiiores específicos de la industria.

    - Por ejemplo, los precios de las acciones en el sector tecnológico pueden verse influenciados por el lanzamiento de nuevos productos , la competencia en el mercado y los avances tecnológicos.

    - Un ejemplo claro es la crisis bancaria de principios de 2023 donde empresas como BBVA , VISA , Mastercard , Santander , etc tuvieron bajas debido a la crisis bancaria iniciada por el Silicon Valley Bank.     


### Una estrategia a largo plazo 

- Un portafolio de inversión siempre es una estrategia a largio plazo y un administradorde portafolios debe ser paciente , resiliente y disciplinado , por lo cual es importante NO CAMBIAR los activos de un portafolio por fluctuaciones de mercado , a menos que:

    - Es momento de rebalancear el portafolio (periodicidad pre-definida).

    - Encuentres un negocio con un potencial de crecimiento más alto.

    - Tu compañía está altamente sobre-valorado , es decir , ya creció hasta su máximo potencial.

    - Si dos años después de la compra la empresa no ha proporcionado rendimientos

    - Si tu activo pasó un umbral de rendimiento pre-definido al armar tu portafolio.

    - Si no puedes soportar la máxima pérdida esperada de tu portafolio (VaR)

    " Speculation is not investment. Always stick to your policies !" - Benjamin Graham

## **Backtesting Dinámico**

### Dynamic Backtesting 

+ Evolución del enfoque estático, ya que incorpora el proceso de rebalanceo en la simulación. Aquí, no solo evaluamos la estrategia inicial, sino también el impacto de los ajustes periódicos que realizamos en la composición del portafolio a lo largo del tiempo.   
  
+ El rebalanceo de un portafolio se realiza de forma sistemática, en intervalos predefinidos, generalmente cada 3, 6 o 12 meses.   
  

+ Si se rebalancea con demasiada frecuencia, los costos de transacción pueden reducir las ganancias, mientras que un rebalanceo poco frecuente puede dejar al portafolio expuesto a desviaciones significativas respecto a su estrategia original.
  

+ Se recomienda realizar la re-optimización de los pesos del portafolio en cada período, utilizando la ventana de datos más reciente. Esto, la vuelve una estrategia de inversión activa al 100%. 
  
+ Este enfoque es más realista porque refleja cómo las decisiones activas del gestor afectan el desempeño bajo diferentes condiciones de mercado. 


### Pasos para hacer un backtesting 


#### 1. Descarga de Datos

Obtener los precios de cierre ajustado en el horizonte temporal recomendado para la estrategia.
    

#### 2. Definir  la ventana de rebalanceo

- Definir la periodicidad "n" del rebalanceo:
    
    - Si se define de manera trimestral (por Q') habrá cuatro rebalanceos por año.
    - Si se define semestral, habrá dos rebalanceos al año.
    - Si se define de manera anual, habrá uno.
     
     
##### 3. Partición de la base de datos
       
- Particionar la base de datos de acuerdo a las "n" ventanas de rebalanceo por año y los "m" años de 
       tiempo disponible.
       
- Tendremos "mxn" subsets -particiones- del mismo data frame.

##### 4. Backtesting
      
- Optimiza los pesos utilizando la primera ventana de tiempo t0.
      
- Simula el rendimiento del portafolio para el período posterior, es decir, de t1 a t2. 
      
- Una vez que finaliza esta ventana, optimiza nuevamente los pesos utilizando datos de t1 a t2 y simula 
      de t2 a t3. 
      
- Repite este proceso de manera iterativa hasta la simulación para el período final, de tn−1 a tn.
      
- Guarda el histórico con la evolución (simulación) del portafolio.
      

##### 5. Visualización de resultados

- Grafica la evolución para comparar el rendimiento de la estrategia a lo largo del tiempo.
       
##### 6. Métricas de Desempeño
        
- Con la evolución histórica del portafolio, obtén las métricas de desempeño.
        
- Analiza las métricas de desempeño y selecciona una estrategia de inversión.
        
- El backtesting dinámico, también nos ayuda a seleccionar la ventana de rebalanceo "óptima" para 
        una estrategia dada.


##### 7. Selección de Estrategia de Inversión

- En base a los resultados, se selecciona una estrategia de inversión y una periodicidad de rebalanceo.
        
- Se obtienen las ponderaciones que se utilizarán para invertir, tomando la última ventana disponible.

## **TAA: Rotación Sectorial**

### Tactical Asset Allocation
- El tactical Asset Allocation es una estrategia activa de gestin de cartera que ajusta el porcentaje de activos mantenidos en diversas categorías para aprovechar las anomalías de precio del mercado a los sectores fuertes del mercado.

- Esta estrategia permite a los gestores de cartera crear valor adicional contra el benchmark aprovechando oportunidades de mercado.

- Supongamos que la asignación de un portafolio tiene 45% acciones , donde 30% son High-Cap y un 15% Low-Cap.

- Si las perspectivas para las acciones de pequeñas empresas no parecen favorables , se pueden ajustar la asignación dentro de las acciones a un 40% HC  y un 5% LC hasta que cambien las condiciones.

- Por lo general los cambios tácticos oscilan entre el 5% - 25% . Se pueden diferenciar dos clases de activos en el TAA: Core Holdings
### Strategic Asset Allocation
- Las "core holdings" son las inversiones fundamentales una cartera de inversión a largo plazo.

- Estas inversiones suelen representar la base central de la cartera y están diseñadas para proporcionar estabilidad y consistencia en rendimientos con respecto al benchmark.

- La idea detras de las "core holdings" es construir un rendimiento estable en la cartera , generalmente a través de inversiones diversificadas.

- Los administradores de portafolio a menudo seleccionan instrumentos financieros como fondos indexados o fondos cotizados en bolsa (ETF's) que replican el rendimiento de un índice específico o una categoría de activos ligadoos a su benchmark.


### TAA / SAA
- Las inversiones tácticas o satélite son aquellas activos que proporcionan oportunidades para el crecimiento adicional de un portafolio de inversión , complementando la estabilidad proporcionada por las inversiones centrales.

- El objetivo de estas acciones tácticas es aprovechar oportunidades de mercado y ayudar a superar el benchmark.


- Una inversión satélite puede ser prácticamente cualquier activo que
un gestor considere que podría ofrecer un potencial adicional en sus
portafolios.

- Esto siempre y cuando el activo cumpla con las políticas de inversión
del portafolio definidos en la “Investment Policy Statement”.

- Se debe establecer una periodicidad para realizar el rebalanceo de
activos tácticos, ya que no es recomendable hacerlo de manera
regular.



### When to be over-under weighted?

Se puede decidir cuando estar overweight o underweight en las "core holdings" , principalmente en función de las siguientes dos estrategias.

- Rotación Sectorial
- Disrupción / Oportunidad en un sector Específico


### *Rotación Sectorial*
### Rotación Sectorial 
- La rotación sectorial es una estrategia de
inversión que consiste en ajustar portafolios
cambiando la asignación de activos entre
sectores según la fase del ciclo económico.

- Se basa en la idea de que el mercado de
valores está influenciado por las condiciones
macroeconómicas, que atraviesan cuatro
ciclos y se repiten de manera iterativa,
además de las expectativas de los
inversionistas.

**Statement:**

- Los mercados de valores tienen altas y bajas
dependiendo de las expectativas económicas que tienen
los participantes.

- Debido a esto, es importante tener en cuenta que hay
momentos donde es conveniente estar “overweight” en
activos agresivos (pro-cíclicos) y ”underweight” en activos
defensivos (anti-cíclicos).



### Beta
- La beta de un portafolio es un indicador que nos indica cuanto se mueve un activo financiero con respecto al mercado, el significado de la beta es el siguiente:
    - β=1: El activo se mueve igual que el mercado.
    - β<1: El activo tiene una relación inversa con el mercado. Por ejemplo, una beta de 0.64 indica que si el mercado sube 1, nuestro activo sube menos (0.64), lo que implica un menor riesgo sistemático.
    - β>1: El activo se mueve en la misma dirección que el mercado, pero con mayor intensidad, lo que conlleva un mayor riesgo sistemático.
### Fases de la economía
- **Macro Bottoming**
    - Earnings Expansion, Rising Equity Markets
        - Condiciones macroeconómicas desfavorables
        - Final del período de contracción
        - Crecimiento de la actividad económica (aumento de la demanda e inversión)
        - Política económica expansiva
        - Acciones infravaloradas con buenos fundamentales, lo que impulsa su precio al alza (el mercado devalores se basa en expectativas y tiende a "predecir" lo que ocurrirá en el ciclo económico)
        - Sentimiento positivo en el mercado bursátil
        - Inversión en activos con beta alta (pro-cíclicos)



- **Macro Improving**
    - Earnings Peak, Equity Markets Peak
        - Mejor momento económico
        - Altas ganancias corporativas
        - Política monetaria restrictiva: Los bancos centrales comienzan a subir las tasas de interés
        - Inicio de período Hawkish
        - Valuaciones elevadas
        - Las acciones alcanzan su techo
        - Sentimiento de euforia en el mercado (todos recomiendan comprar)
        - Se recomienda comenzar transición de activos pro-cíclicos a anti-cíclicos


- **Macro Peaking**
    - Earnings Peak Out, Equity Markets Falling
        
        - Crecimiento económico insostenible y contracción de la demanda
        - Disminución de las ganancias empresariales
        - Tasas de interés en máximos históricos
        - Caída en los mercados bursátiles
        -Sentimiento pesimista en el mercado
        - Se recomienda tener el portafolio compuesto por activos defensivos (Beta baja y empresas en
        sectores inelásticos)


- **Macro Declining**
    - Earnings Decline, Equity Markets bottoming

        - Comienzo del período de recesión
        - Caída de las ganancias empresariales
        - Política económica expansiva: Inician recortes en las tasas de interés
        - Comienzo de período Dovish
        - El mercado de valores alcanza su punto más bajo, los fundamentales comienzan a mejorar
        - Sentimiento de mercado mixto
        - Transición hacia activos más agresivos



   

### Hawkish
- Apreciación de la Moneda
- Alza de Tasas de Interés
- Política Monetaria Restrictiva
- Se busca reducir la inflación

### Dovish
- Depreciación de Moneda
- Baja de Tasas de Interés 
- Política Monetaria Expansiva
- Se busca estimular el crecimiento económico

### *Sectores y Cíclos Económicos*

### Rotación de Activos
- Para saber cómo rotar dentro de las inversiones de
una cartera, es importante conocer la naturaleza
pro-cíclica o anti-cíclica de un activo o sector, para
poder separarlos sectorialmente.

- Cabe destacar, que existen tres momentos
importantes en el ciclo de una cartera (Underweight,
Overweight y Neutral) y estos tienen que ser
definidos en el prospecto de inversión.
### Sectores Pro-Ciclícos
- Los activos pro-cíclicos son acciones de empresas cuyos ingresos y beneficios están fuertemente
influenciados por el ciclo económico.

- Estas empresas operan en sectores donde los resultados financieros son más susceptibles a
fluctuaciones influenciadas por variaciones en la demanda del consumidor y las condiciones
macroeconómicas. Son activos elásticos por naturaleza y con una β > 0.7 .

- Algunos ejemplos de sectores pro-cíclicos:

    - Consumer Discretionary
    - Tech
    - Real State
    - Industrial & Communications Sector
    - Financial
### Sectores Anti-Ciclícos 

- Los activos anti-cíclicos son aquellas empresas o bienes que tienden a mantener un performance relativamente estable o incluso mejor durante recesiones económicas.

- Esto se debe a que son productos esenciales o con una demanda constante (bienes inelásticos),
independientemente de las condiciones económicas. Son activos inelásticos por naturaleza y con una
0<β<0.7. Los activos “refugio” son históricamente activos anti-cíclicos.

- Algunos ejemplos de sectores anti-cíclicos:
    - Consumer Staples
    - Healthcare
    - Utilities
    - Safe-Haven Commodities

### *Leading Indicators*

### Composite Leading Indicator

- El CLI es un indicador diseñado para proporcionar
señales tempranas de puntos de inflexión y
cambios en la tendencia en ciclos económicos,
mostrando la fluctuación esperada de la actividad
económica basandose en datos.


- Se basa en estimados económicos, como la
actividad económica, el producto interno bruto,
consumo, comercio doméstico e internacional,
política monetaria y fiscal.
Este indicador esta normalizado a base 100.


### Business Confidence Index

- El BCI, es un indicador de confianza
empresarial que proporciona información
sobre la opinión y confianza de las
empresas sobre la economía.
Puede utilizarse para monitorear el
crecimiento de la producción y anticipar
puntos de inflexión en la actividad
económica.

- Números por encima de 100 sugieren
mayor confianza.
- Números por debajo de 100 indican
pesimismo.

- Este indicador suele ser utilizado como una medida
para determinar las expectativas de las empresas
con respecto al crecimiento económico.

- Suele ir históricamente adelantado a las
fluctuaciones del S&P 500 (En términos relativos).

    
### Consumer Confidence Index
- El CCI proporciona información sobre el comportamiento
esperado en el consumo y el ahorro de los hogares, basado
en encuestas relacionadas con la situación financiera
esperada, percepción sobre la situación económica, el
desempleo y la capacidad de ahorro de las personas.

- Valores por encima de 100 señalan confianza de los
consumidores hacia la situación económica futura, como
consecuencia son menos propensos a ahorrar y más
inclinados a gastar dinero.

- Valores por debajo de 100 indican una actitud pesimista
hacia la situación económica, posiblemente resultando en
una tendencia a ahorrar más y consumir menos.

- Este indicador suele ser utilizado como
una medida para determinar las
expectativas del público con respecto al
crecimiento económico.

## **Alpha: Eficiencia del Mercado Bursátil**


### Alpha 
- La Alpha, como vimos en clases previas, es la tasa de rendimiento
por arriba del rendimiento de un mercado que un administrador de
portafolios genera en un período.

- También es conocido como el “rendimiento activo” de una inversión
y el complemento (rendimiento del mercado) es conocido como
“rendimiento pasivo”.
- La alfa es importante ya que nos indica dos cosas:
    1. Que tan bueno es un administrador de portafolios. ¿Cuánto le pagamos?
    2. ¡Un alfa positivo indica ineficiencias de mercado!

### Eficiencia del Mercado 
- Recordando, uno de los supuestos del CAPM era:
    - Todos los inversionistas tienen acceso a la misma
información, por lo tanto arman sus carteras de una
manera idéntica.

- La eficiencia de mercados nos indica que TODA la
información disponible se encuentra en el precio de un
activo de manera insesgada.

- Si es posible generar un Alpha, nos encontramos con
una discrepancia en esta teoría.

### Hipótesis del Mercado 
- Desarrollada por Eugene Fama en la década de 1970.

- La EMH explica que cada mercado tiene características
propias, y dependiendo de la cantidad de inversionistas, los
costos de transacción, la disponibilidad y la velocidad de la
información y las restricciones al operar se puede determinar
la eficiencia de un mercado concreto,

- Se identifican tres formas de eficiencia:
    - **Weak-Form Market Efficiency**
        - Los precios reflejan toda la información pasada del mercado (precios y volumen).
        - El análisis técnico no permite obtener rendimientos superiores.
        - Los precios siguen una caminata aleatoria.


    - **Semi-Strong Market Efficiency**
        - Los precios reflejan toda la información pública disponible (incluye noticias, estados financieros, etc.).
        - El análisis fundamental no genera rendimientos ajustados por riesgo positivos en promedio.
        - El mercado reacciona rápidamente a nueva información pública (inesperada).


    - **Strong Market Efficiency**
        - Los precios reflejan toda la información, pública y privada.
        - Nadie puede obtener “abnormal returns” de forma consistente, ni siquiera con información privilegiada.
        - En la práctica, los mercados no son eficientes en su forma fuerte, debido a la prohibición del “inside
        trading”.


### Ineficiencia de Mercados 

- Una ineficiencia de mercado se refiere a la discrepancia
entre el precio y el valor de un activo. Cuando se dice que un
activo cotiza “a descuento” significa que su precio es menor
que su valor real.
- Al identificar y explotar estas ineficiencias puedes generar
Alfa.

- Las discrepancias de mercado pueden surgir debido a
diversas razones, como información asimétrica,
comportamientos irracionales de los inversionistas,
limitaciones en el acceso a ciertos activos o segmentos del
mercado.


- En la bolsa, entre 80% y 90% de las personas que invierten en bolsa pierden dinero
anualmente, debido a desinformación, comportamiento y sesgos cognitivos.
¡Evita caer en estos!


### Desinformación 

- Falta de información precisa, completa o
confiable sobre un activo financiero o sobre el
mercado en general.

- Puede surgir por distintos factores como:
    - Rumores y Fake News
    - Manipulación de Mercado
    - Informes engañosos y poco claros.


### Comportamiento Irracional 

- Decisiones de inversión o acciones
que se toman basadas en emociones,
prejuicios o creencias irracionales en
lugar de fundamentos o análisis
cuantitativo.

- Algunos ejemplos:
    - Euforia y Exceso de Confianza
    - Miedo y Pánico
    - Mentalidad de Rebaño



### Sesgos Cognitivos 

- Un sesgo cognitivo es un patrón sistemático de desviación en el pensamiento o la percepción que
puede llevar a la toma de decisiones irracionales o sesgadas.

- En finanzas bursátiles, la rama que estudia estos patrones se denomina “behavioral finance”,
algunos de estos sesgos son:

- Sesgo de confirmación: Tendencia a buscar información que respalde sus creencias existentes
y descartar información que las contradiga. Esto puede llevar a la interpretación selectiva de
datos y a decisiones de inversión sesgadas, ignorando información relevante que podría
afectar la evaluación de los activos.

- Sesgo de aversión a las pérdidas: Los inversionistas valoran más evitar pérdidas que obtener
ganancias.

- Sesgo de anclaje: El sesgo de anclaje ocurre cuando los inversionistas se basan
excesivamente en un valor o punto de referencia inicial al tomar decisiones.


### Toma de Decisiones Informadas 

- El asset allocation cuantitativo es una estrategia que puede ayudar a identificar y aprovechar
las ineficiencias de mercado, al investigar la coyuntura y con esta información, hacer uso de
modelos cuantitativos para asignar de manera óptima los recursos entre diferentes clases de
activos o instrumentos financieros.

## **Operación en el Mercado Bursátil**


### Spot Price 

- Precio actual en el mercado para la compra
o venta de un activo. Se determina por la
oferta y la demanda en el mercado,
reflejando las condiciones actuales de
operación

### Precio de Cierre
- Último precio al que se negocia un activo
financiero al final de la sesión de mercado, en la
BMV, se determina por el precio promedio
ponderado de los últimos 20 minutos de
operación.


### ASK & BID Price
- **BID**
    - Precio más alto al que un inversor esta dispuesto a comprar un activo financiero.

- **ASK**
    - Precio más bajo al que un inversor esta dispuesto a VENDER un activo financiero

### SPREAD 

- El Spread es la diferencia entre el Ask y el Bid.

- El Spread nos da una buena idea de que tan líquido
es un activo financiero, entre más pequeño es, más
fácil es comprarlo y venderlo.

- Es por esto, que el spread puede ser explicado
como un costo de transacción, que también puede
incluir la comisión que te cobra el exchange por
operar, un “costo de obtener liquidez”.


### Riesgo de Liquidez de Mercado 

- El riesgo de liquidez de mercado es el riesgo no diversificable de
que una posición financiera no pueda ser fácilmente liquidada
en el corto plazo sin afectar significativamente su valor debido a
la falta de profundidad de mercado.

- Esto ocurre cuando no hay suficientes compradores o
vendedores en el mercado para acomodar grandes
transacciones sin causar fluctuaciones sustanciales en los
precios.

- En consecuencia, el riesgo de liquidez está directamente
relacionado con la eficiencia del mercado.


### Mercados NO Líquidos

- En mercados eficientes, los participantes pueden ejecutar
transacciones de manera fluida, sin generar grandes movimientos
en los precios, lo que les permite comprar o vender activos
rápidamente y con costos de transacción mínimos.

- Sin embargo, en mercados ineficientes, donde la liquidez es
escasa, incluso pequeñas transacciones pueden provocar
fluctuaciones significativas en los precios.

- Esto aumenta los costos de transacción y puede obligar a los
inversores a aceptar precios menos favorables, lo que en última
instancia puede deteriorar el rendimiento general de un portafolio
de inversión.


### Volumen 

- Número total de títulos de un activo financiero
negociados durante un período específico,
generalmente a lo largo de una sesión de
mercado.

- Refleja la actividad de compraventa de un activo
en particular y es utilizado para medir el interés de
los inversores y la liquidez.

- Un alto volumen indica una mayor actividad y, a
menudo, mayor liquidez, lo que facilita la
ejecución de transacciones sin generar
fluctuaciones significativas en el precio.


### Días para liquidar una posición 

- Aunque el riesgo de liquidez es complejo de cuantificar, ya que depende del valor total del
portafolio y de la liquidez de sus activos constituyentes, existen diversas metodologías para
estimarlo.

- La convención del mercado es tomar la proporción entre el número de acciones que se tienen y
el volumen promedio diario (ADV, por sus siglas en inglés) de los últimos tres meses para medir la
liquidez de un activo. Esta métrica estima cuántos días serían necesarios para liquidar (vender)
completamente la posición en dicho activo. La fórmula es la siguiente:


$$N_{days} = \frac{N_{shares}}{ADV}$$

- Donde:
    - N shares: Número de acciones de un activo contenidas en un portafolio
    - ADV: Volumen promedio diario del activo en los últimos tres meses


### Volumen : Niveles 
- Los niveles de un activo financiero reflejan el interés en
volumen para cada precio de compra y venta, antes de que
se ejecute la operación.

- Esta medida indica el volumen al que los participantes del
mercado están dispuestos a comprar o vender un activo,
así como los precios correspondientes.

- Los niveles de precio se utilizan para tomar decisiones de
compra o venta. Por ejemplo, algunos inversionistas
emplean niveles de precios específicos como puntos
estratégicos de entrada o salida en una posición.


### Volumen : Profundidad 

- La "profundidad" de un activo financiero hace referencia a
la cantidad de órdenes de compra y venta disponibles a
diferentes precios en un mercado.

- En otras palabras, mide cuántas órdenes existen para un
activo a distintos niveles de precios.

### Órdenes de Compra y Venta 

- Instrucción o solicitud que un inversor envía a su
intermediario financiero para comprar o vender un
activo financiero.

- Una orden establece los detalles clave de la
transacción, como el tipo de orden, el activo a
negociar, la cantidad, el precio.

- Las órdenes son utilizadas por los inversores para
ejecutar transacciones en los mercados financieros
de manera estructurada y precisa.

- El spread máximo en GBM se limita a una variación
del 5% respecto al precio spot.

### Tipos de Órdenes 

- A mercado (MKT): Instrucción para comprar o vender un activo al mejor precio
disponible en el mercado en el momento de ejecución.

- Limitada: Instrucción para comprar o vender un activo a un precio específico o
uno mejor.

- Volumen Oculto: Instrucción de compra o venta en la que el inversor solicita
ejecutar una transacción sin revelar públicamente la cantidad total que desea
comprar o vender. También se conoce como "iceberg order". Es funcional
cuando tienes altos volúmenes de transacción (Market Makers, Fondos) para no
mover el precio.

- Stop: Es una instrucción para comprar o vender un activo cuando este alcanza
un precio específico conocido como "precio de activación". Una vez que se
alcanza ese precio, la orden se convierte en una orden de mercado (Stop en
línea) o limitada (Stop limitado), dependiendo de la configuración establecida.


### Tipos de Estructuras de Mercado 

- Orden-driven markets
    - Las órdenes se ejecutan mediante
reglas de negociación en bolsas o
sistemas automáticos.
    - Orden de prioridad: mejor precio y
llegada más temprana.
    - Ejemplo: redes electrónicas (crossing
networks)


- Quoted-driven markets
    - Operaciones se realizan con dealers o market
    makers, que publican precios de compra y
    venta.
    - También llamados mercados OTC o de
    dealers.
    - Usuales en mercados de renta fija o
    derivados.

- Brokered-driven markets
    - Brokers buscan la contraparte ideal para
    ejecutar la operación.
    - Útiles para activos ilíquidos o únicos (bloques
    grandes de acciones, bienes raíces, arte,
    private capital).
    - No hay transacciones en el mercado
    tradicional.

## **Investment Policy Statement**

### Contexto 

- Es muy poco probable que un gestor de carteras pueda
producir resultados adecuados para un cliente sin conocer sus
características y preferencias, por tanto, siempre que se da
asesoría y gestión de inversiones se debe de generar un
investment policy statement (IPS).

- El objetivo del IPS es:
    - Tener un documento escrito que sirva como referencia
    para la toma de decisiones objetivas.
    - Establecer objetivos de inversión, tolerancia al riesgo y
    restricciones específicas.
    - Proporcionar claridad sobre las necesidades y
    circunstancias del cliente.
    - Alinear expectativas de riesgo y retorno.


### Componentes Principales del IPS

1. Propósito del IPS.
2. Responsabilidades de las partes involucradas.
3. Procedimientos de actualización y respuesta a situaciones.
4. Objetivos de inversión (RR).
5. Restricciones de inversión (TTLLU).
6. Guías de inversión: tipos de activos permitidos y uso de
apalancamiento.
7. Evaluación del desempeño: benchmarks y métricas.
8. Apéndices: Strategic asset allocation y políticas de
rebalanceo.



### Evaluación de la Aversión al Riesgo

- Antes de determinar los objetivos RR (Risk & Return) del cliente
se debe de evaluar la aversión al riesgo de este, tomando en
cuenta los siguientes dos factores:

- Client ability to bear risk:

    - Depende de sus circunstancias financieras y
personales, por ejemplo:
        
        - Una persona joven tiene horizontes de inversión
        más largos.
        - Patrimonio en relación a los gastos.
        - Seguros ante eventos inesperados.
        - Estabilidad del empleo.


- Client willingness to bear risk:
    
    - Se basa principalmente en las actitudes y
creencias sobre las inversiones.
    
    - La evaluación de la actitud de un
inversionista hacia el riesgo es bastante
subjetiva y, a veces, se realiza mediante un
breve cuestionario que intenta categorizar
la aversión o tolerancia al riesgo del
inversionista.


### Objetivo de Riesgo y Rendimiento 
- Una vez establecido el nivel de aversión al riesgo del cliente, se
puede pasar a determinar los objetivos RR (Riesgo y Rendimiento del
mismo).

- El riesgo que el cliente puede asumir el cliente determina el nivel
de rendimiento que este puede llegar a alcanzar, por ejemplo:
    
    - Aversión al riesgo baja: Invertir en activos de renta fija, poco
    riesgo, rendimiento esperado de 5% aproximadamente.
    
    - Aversión al riesgo media: Invertir en activos de renta fija y
    renta variable, rendimiento esperado entre 8-12%.
    
    - Aversión al riesgo alta: Invertir en renta fija, variable,
    apalancar posiciones y activos alternativos, rendimiento
    esperado del 12-15%.


- Objetivos de riesgo:
    - Absolutos: evitar pérdidas mayores al X% en un período determinado.
    - Relativos: mantener el rendimiento dentro de un rango respecto a un
    benchmark.

- Objetivos de retorno:
    - Nominales: lograr un retorno anual del X%.
    - Reales: superar la inflación en X%.
    - Relativos: exceder el rendimiento de un índice de referencia en X%.


### Restricciones de Inversión 

- Las restricciones de inversión cuentan con cinco factores
específicos (TTLLU):

    - Tax Constraints: Impacto de impuestos en las inversiones.

    - Time Horizon Contraints: Duración prevista de la inversión.
    
    - Liquidity Constaints: Necesidades de efectivo a corto
    plazo.
    
    - Legal & Regulatory Constraints: Restricciones impuestas
    por leyes o regulaciones.
    
    - Unique Circumstances: Preferencias éticas, restricciones
    específicas o necesidades particulares.


### Definir Clases de Activos Permitidos 

- Definir clases de activos en los cuales se esta permitido invertir,
tomando en cuenta que:
    
    - Los activos/sectores deben de tener una alta correlación
    interna y baja correlación entre activos para lograr una buena
    diversificación.
    
    - Ejemplos: acciones nacionales/internacionales, bonos
    gubernamentales/corporativos, bienes raíces, inversiones
    alternativas.
    
    - La selección debe alinearse con los objetivos y restricciones
    del IPS.


### Asset Allocation

- Establecer la metodología con la cual se administrará el
portafolio, por pasos:

    - Establecer si se seguirá una política pasiva o activa.
    
    - En caso de que sea activa, definir el método que se
    seguirá para optimizar el portafolio, además de en que
    casos puede cambiar.
    
    - Definir el Strategic Asset Allocation.
    
    - Definir el Tactical Asset Allocation, y en que casos el
    gestor puede cambiar discrecionalmente las
    asignaciones.

### Performance Evaluation 

- Definir los principios para evaluar al gestor, estableciendo:

    - El benchmark contra el cual se comparará el resultado del
    portafolio.

    - Establecer las métricas con las que se evaluará el portafolio,
    como retorno absoluto, retorno relativo, Sharpe, Jensen, etc.

    - Establecer la frecuencia con la cual se realizará el reporte de
    inversión (al menos, debe ser trimestral)



### Políticas de Rebalanceo (Apéndice)

- Establecer políticas claras sobre cuándo y cómo rebalancear.

- Considerar:
    - Rebalanceo periódico o sistemático.
    - Rebalanceo ante cambios estructurales: Umbrales de
    desviación permitidos.

- El objetivo del rebalanceo mantiene la alineación con los
objetivos del IPS, por tanto, en caso de que se modifiquen las
ponderaciones objetivo, independiente del TAA establecido,
se debe modificar la IPS.

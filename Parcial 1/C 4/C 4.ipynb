```python
prices = yf.download(tickers, start='2021-01-01')['Close']
```

```python
returns = prices.pct_change().dropna()
```

```python
daily_returns= (prices/prices.shift()-1).dropna()
```

```python
df= pd.DataFrame(columns=returns.keys()).T
### Returns
df['Daily_rate']= returns.mean()
df['Anual_rate']= returns.mean()*252

### Volatility / desv 
df['Daily_Vol']= returns.std()
df['Anual_Vol']= returns.std()*np.sqrt(252)

### Coeficiente de Variación
df['CV']= df['Anual_Vol']/df['Anual_rate']

### Asimetria
df['Skewness']= returns.skew()

df
```

```python
sorted_df= df.sort_values(by='CV')
```

```python
sorted_df[sorted_df['CV'] > 0].head(10)
```

```python
df= pd.DataFrame(columns=returns.keys()).T
### Returns
df['Daily_rate']= daily_returns.mean()
df['Anual_rate']= daily_returns.mean()*252

### Volatility / desv 
df['Daily_Vol']= daily_returns.std()
df['Anual_Vol']= daily_returns.std()*np.sqrt(252)

### Coeficiente de Variación
df['CV']= df['Anual_Vol']/df['Anual_rate']

### Asimetria
df['Skewness']= daily_returns.skew()

df
```

```python
w= np.ones(6)/6
w
```

```python
Rend_port = sum(df['Anual_rate'] * w)
Rend_port*100
```

```python
COV = daily_returns.cov()

COV
```

```python
Var_port = (w.T @ COV @ w) * 252
Volat_port = np.sqrt(Var_port)
Var_port*100, Volat_port*100
```
